####Los tipos de variables se determinaron para este trabajo practico de la siguiente manera:

* **Cualitativas o Categóricas**:
Estas variables representan cualidades, categorías o etiquetas que no pueden medirse con números. No tienen una magnitud numérica inherente y describen características que no pueden cuantificarse directamente. Se dividen en dos subtipos:

* **Nominales**: Las categorías no tienen un orden inherente. Ejemplos: género (masculino, femenino), color de ojos (azul, marrón, verde).

* **Ordinales**: Las categorías tienen un orden o jerarquía. Ejemplos: nivel educativo (primario, secundario, universitario), clasificación socioeconómica (bajo, medio, alto).

* **Cuantitativas o Numéricas**:
Estas variables se expresan en términos de números y tienen una magnitud numérica significativa. Se dividen en dos subtipos:

* **Discretas**: Los valores son números enteros y tienen un recuento finito o numerable. Ejemplos: número de hijos, número de libros en una biblioteca.

* **Continuas**: Los valores pueden ser números reales en un rango específico. Pueden tomar cualquier valor dentro de un intervalo. Ejemplos: altura, peso, temperatura.

####Dataset wine
El presente dataset está relacionado con las variantes roja y blanca del vino "Vinho Verde" portugués.
 Es un dataset popular en el campo del análisis de datos y el machine learning. Contiene información sobre las propiedades físico-químicas de diferentes tipos de vinos, así como calificaciones de calidad asignadas por catadores. Este conjunto de datos es comúnmente utilizado para ejercicios de clasificación y regresión, y es un buen ejemplo para prácticas de análisis de datos y construcción de modelos predictivos.

**Las variables que analiza son:**
1. **Fixed Acidity** (Acidez Fija): La cantidad de ácido tartárico presente en el vino. Es un factor clave para determinar la acidez total del vino.**Variable Numerica Continua**

2. **Volatile Acidity** (Acidez Volátil): La cantidad de ácido acético en el vino, que puede contribuir a un sabor desagradable si está presente en exceso.**Variable Numerica Continua**

3. **Citric Acid** (Ácido Cítrico): La cantidad de ácido cítrico en el vino. Puede aportar frescura y sabor cítrico al vino.**Variable Numerica Continua**

4. **Residual Sugar** (Azúcar Residual): La cantidad de azúcar que queda en el vino después de que se complete la fermentación.**Variable Numerica Continua**

5. **Chlorides** (Cloruros): La cantidad de sales de cloruro en el vino. Puede afectar el sabor y la sensación en boca.**Variable Numerica Continua**

6. **Free Sulfur Dioxide** (Dióxido de Azufre Libre): La cantidad de dióxido de azufre disuelto en el vino que se encuentra en forma libre, no unido a otras moléculas.**Variable Numerica Continua**

7. **Total Sulfur Dioxide** (Dióxido de Azufre Total): La suma del dióxido de azufre libre y el dióxido de azufre unido a otras moléculas en el vino.**Variable Numerica Continua**

8. **Density** (Densidad): La densidad del vino, que puede estar relacionada con la cantidad de alcohol y azúcar en el vino.**Variable Numerica Continua**

9. **pH**: El nivel de acidez o alcalinidad del vino.**Variable Numerica Continua**

10. **Sulphates** (Sulfatos): La cantidad de sales de sulfato en el vino. Puede contribuir al sabor y la estabilidad del vino.**Variable Numerica Continua**

11. **Alcohol**: El contenido de alcohol en el vino.**Variable Numerica Continua**

12. **Quality** (Calidad): La variable objetivo del conjunto de datos. Representa la calidad del vino, puntuada en una escala del 0 al 10 por catadores.**Variable Numerica Discreta**

13. **Style** (Estilo): Variable que diferencia entre Tinto y blanco los vinos muestreados.**Variable Categorica**

```python
def detect_outliers_iqr(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers
```

```python
df=df_diabetes
columns_to_check = ["Pregnancies",	"Glucose",	"BloodPressure",	"SkinThickness",	"Insulin",	"BMI",	"DiabetesPedigreeFunction",	"Age"]
outliers_list = []
for col in columns_to_check:
    outliers = detect_outliers_iqr(df, col)
    outliers_list.append(outliers)

diab_outliers_df = pd.concat(outliers_list)
diab_outliers_df
```

Una vez trabajadas las columnas del dataframe se pasara a parsear los datos para estandarizar los tipos de datos que el dataset maneja.

```python
def replace_values(df, column_name, old_value, new_value):
  """
  Reemplaza los valores de una columna en un DataFrame.
  Args:
    df (pd.DataFrame): El DataFrame en el que se reemplazarán los valores.
    column_name (str): El nombre de la columna en la que se reemplazarán los valores.
    old_value (any): El valor antiguo que se reemplazará.
    new_value (any): El nuevo valor que reemplazará al antiguo.
  Returns:
    pd.DataFrame: El DataFrame con los valores reemplazados.
  """
  # Comprueba si la columna existe en el DataFrame.
  if column_name not in df.columns:
    raise ValueError(f"La columna {column_name} no existe en el DataFrame.")
  # Reemplaza los valores antiguos por los nuevos.
  df[column_name] = df[column_name].replace(old_value, new_value)
  # Devuelve el DataFrame con los valores reemplazados.
  return df
```

```python
replace_values(df_titanic_final,"survived",1,True)
replace_values(df_titanic_final,"survived",0,False)
replace_values(df_titanic_final,"alive","no",False)
replace_values(df_titanic_final,"alive","yes",True)
```
